##### 2. display the continent where positivity rate and the strigency index is maximum

```
%sql
SELECT continent,MAX(stringency_index) AS si,MAX(positive_rate)
FROM covid
GROUP BY continent
```

##### 3.which continent and Country having max of deaths and male_smokers

```
%sql
select continent,country,MAX(new_deaths),MAX(male_smokers)
from covid
group by continent,country
```

##### 4. which year having the highest number of deaths

```
%sql
SELECT date,sum(new_deaths) AS Total_Death
from covid
group by date
```

##### 5. Countries with max cardiovasc_death_rate and diabetes_prevalence greater than 10

```
%sql
SELECT country,MAX(cardiovasc_death_rate)
from covid
where diabetes_prevalence >10
group by country
```

##### 6.which location has highest number of vaccinated people in each continent

```
%sql
select continent,country,max(new_vaccinations) AS Highest_num_vaccinated
from covid
group by continent,country
```
